Read the dataset from the below link
https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv', sep=',')
print(type(df))
df.head(15)
```

1. Delete unnamed columns

```python
del df['Unnamed: 0']
```

2. Show the distribution of male and female

```python
df['Gender'].describe()
```

3. Show the top 5 most preferred names

```python
df['Name'].value_counts(normalize=True, sort=True, ascending=False, bins=None)
```

4. What is the median name occurence in the dataset

```python
df['Name'].loc[df['Id']==df['Id'].median()]
```

5. Distribution of male and female born count by states

```python
df.groupby('State')['Gender'].describe()
```
